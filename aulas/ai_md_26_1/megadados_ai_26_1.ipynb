**Exercício 3**: A plataforma deseja analisar o engajamento dos usuários que cadastram locais de acampamento. Crie uma query que retorne:

- O nome de usuário (coluna `nome_usuario`);
- O e-mail do usuário;
- O número total de locais cadastrados pelo usuário (coluna `qtde_locais`);
- A nota média das avaliações recebidas nos locais cadastrados pelo usuário (coluna `nota_media`).

**Obs**:

- Usuários que não cadastraram nenhum local também devem ser retornados com valores zerados (onde se aplica).
- A nota média deve considerar todas as avaliações de todos os locais cadastrados pelo usuário.
- Ordene de forma decrescente pela quantidade de locais. Em caso de empate, ordene de forma crescente pelo nome de usuário.

````python
sql_ex03 = """
SELECT u.nome_usuario, u.email, COUNT( DISTINCT la.id_local) as qtde_locais, IFNULL(AVG(a.nota), 0) as nota_media
FROM usuario u
LEFT JOIN local_acampamento la 
ON u.id_usuario = la.id_usuario_cadastro
LEFT JOIN avaliacao a USING(id_local)
GROUP BY u.id_usuario
ORDER BY qtde_locais DESC, u.nome_usuario ASC
"""

db(sql_ex03)
````

**Exercício 4**: A equipe de marketing deseja analisar a popularidade dos locais de acampamento por país.

Crie uma view `v_relatorio_paises` que retorne (nesta ordem):

- O nome do país (coluna `nome_pais`);
- A sigla do país (coluna `sigla`);
- O número total de locais de acampamento no país (coluna `qtde_locais`);
- O número total de avaliações nos locais do país (coluna `qtde_avaliacoes`);
- A nota média das avaliações dos locais do país (coluna `nota_media`);
- O preço mínimo da diária dos locais do país (coluna `preco_minimo_diaria`);
- O preço máximo da diária dos locais do país (coluna `preco_maximo_diaria`).

**Obs**:
- Retorne apenas países que possuam pelo menos 2 locais de acampamento.
- Ordene de forma decrescente pela nota média. Em caso de empate (não haverá, mas faça!), ordene de forma crescente pelo nome do país.
- Utilize apenas uma query para criar a view.
- Não envie comando de `DROP` para o servidor.
- Caso algum país possua dois locais ou mais, mas os locais não possuam avaliações, considere a `qtde_avaliacoes` e a `nota_media` como zero.
- Considere apenas a coluna `preco_diaria` para os cálculos de preço mínimo e máximo. Este é o preço padrão para os cálculos, mesmo que existam outras colunas de preço (ex: `preco_diaria_baixa_temporada` e `preco_diaria_alta_temporada`).

````python
sql_ex04 = """
CREATE OR REPLACE VIEW v_relatorio_paises AS
    SELECT 
        p.nome_pais,
        p.sigla,
        COUNT( DISTINCT la.id_local) AS qtde_locais,
        COUNT(a.id_avaliacao) AS qtde_avaliacoes,
        COALESCE(AVG(a.nota), 0) AS nota_media,
        MIN(la.preco_diaria) AS preco_minimo_diaria,
        MAX(la.preco_diaria) AS preco_maximo_diaria
    FROM pais p
    LEFT JOIN cidade c USING(id_pais)
    LEFT JOIN local_acampamento la USING(id_cidade)
    LEFT JOIN avaliacao a USING(id_local)
    GROUP BY p.nome_pais, p.sigla
    HAVING qtde_locais >= 2
    ORDER BY nota_media DESC, p.nome_pais ASC
"""

db(sql_ex04)
````
